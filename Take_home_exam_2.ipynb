**Question 5:** Compare the predictive accuracy of a multiple linear regression model versus a lasso model when predicting `rented_bike_count`. Use all other variables as feature variables. Include all necessary steps of ML workflow and use the appropriate accuracy metric. Which model does better?

```python
# Target & features declaration
target = ride_share['rented_bike_count']
features = ride_share.loc[:, ['temperature', 'humidity', 'wind_speed', 'visibility',
             'dew_point', 'solar_radiation', 'rainfall', 'snowfall',
                              'holiday', 'seasons', 'month', 'hour', 'normal_day']]


# Identifying categorical and numerical features
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import make_pipeline

encoder = OneHotEncoder(sparse_output=False)

# I am including month & hour as categorical features because the relationship
# between them and bike rentals is non-linear and has distinct differences between each periods
cat_feats = ['holiday', 'seasons', 'normal_day', 'month', 'hour']
num_feats = ['temperature', 'humidity', 'wind_speed', 'visibility',
             'dew_point', 'solar_radiation', 'rainfall', 'snowfall']

# Creating ColumnTransformer for preprocessing
final_pipe = ColumnTransformer([
    ('cat', encoder, cat_feats)
])

# Preprocess training and test data
features_processed = final_pipe.fit_transform(features)

# Splitting dataset into training and testing sets
from sklearn.model_selection import train_test_split
feat_train, feat_test, tgt_train, tgt_test = train_test_split(features_processed, target,
                                                              test_size=0.3, random_state=42)
print("Dimension for training features: ", feat_train.shape)
print("Dimension for testing features: ", feat_test.shape)


# **Linear Regression**
from sklearn.linear_model import LinearRegression, Lasso
model = LinearRegression()
model.fit(feat_train, tgt_train)

# Predictions and performance
from sklearn.metrics import mean_squared_error
predicted_rented_bike = model.predict(feat_test)


# Compare predicted vs actual

predicted_vs_actual = pd.DataFrame({
        'actual_rented_bike' : list(tgt_test),
        'predicted_rented_bike' : list(predicted_rented_bike)
    })
print(predicted_vs_actual.head(n = 10))

# Calculate MSE
mse = mean_squared_error(tgt_test, predicted_rented_bike)

print(f"Linear Regression MSE: {mse:.4f}")


# **Lasso Regression**
lasso_model = Lasso(alpha=1.5, max_iter=100000)
lasso_model.fit(feat_train, tgt_train)

# Predictions and performance
lasso_predictions = lasso_model.predict(feat_test)
lasso_mse = mean_squared_error(tgt_test, lasso_predictions)

print(f"Lasso Regression MSE: {lasso_mse:.4f}")
```

-
Based on the Mean Squared Error values, Linear Regression is the better model in this case, as it has a lower MSE, 162938.3851 compared to Lasso Regression 163305.9623. This indicates that Linear Regression has better predictive accuracy for predicting rented_bike_count in this scenario and regularization via Lasso might be causing the model to underfit by shrinking coefficients too much.


- N.B: I checked the models while including both month and hour as numerical features, including one of them as numerical and the other as catogorical. Including them as categorical yielded the lowest MSE for both Lasso and Linear regression model. I believe it is because these two variables have a none linear relationship with rented bike count. **Question 6:** Build a lasso model by tuning the `alpha` hyper-parameter. Use a 5 fold cross validation to do this. Compare the predictive accuracy of the tuned lasso model with the non-tuned lasso model and the linear regression model in question 5. Which model does better and why?

```python
# **Lasso Model Hyperparameter Tuning**
from sklearn.model_selection import train_test_split, cross_val_score, KFold
alpha_values = [0.001, 0.01, 0.1, 1, 10, 100, 1000, 10000, 100000]
kf = KFold(n_splits=5, shuffle=True, random_state=42)

alpha_mse = []

for alpha in alpha_values:
    model = Lasso(alpha=alpha, max_iter = 100000)
    mse_scores = cross_val_score(model, feat_train,
                                 tgt_train, cv=kf, scoring='neg_mean_squared_error')
    mean_mse = - np.mean(mse_scores)
    alpha_mse.append(float(mean_mse))

# Plotting MSE vs. Alpha
plot_data = pd.DataFrame({"alpha": alpha_values, "mse": alpha_mse})

(
    ggplot(plot_data, aes(x="alpha", y="mse")) +
    geom_line() +
    geom_point() +
    labs(
        title="MSE vs. Alpha for Lasso Regression",
        x="Alpha",
        y="Mean Squared Error"
    )
)
```

```python

# Selecting the best alpha value
best_alpha_index = np.argmin(alpha_mse)
best_alpha = alpha_values[best_alpha_index]
print(f"Best alpha value: {best_alpha}")

# **Tuned Lasso Model**
lasso_tuned = Lasso(alpha=best_alpha, max_iter=100000)
lasso_tuned.fit(feat_train, tgt_train)

# Predictions and performance for tuned Lasso
predicted_lasso_tuned = lasso_tuned.predict(feat_test)
mse_lasso_tuned = mean_squared_error(tgt_test, predicted_lasso_tuned)

print(f"Tuned Lasso MSE: {mse_lasso_tuned:.4f}")
```

The Linear Regression model achieved an MSE of 162938.3851, performing better than the non-tuned Lasso model, which had an MSE of 163305.9623. The non-tuned Lasso model likely underperformed due to a relatively large alpha value, causing underfitting and limiting its ability to capture important patterns in the data.

However, the tuned Lasso model, with an MSE of 162587.0749, outperformed both Linear Regression and non-tuned Lasso. By using 5-fold cross-validation to optimize the alpha parameter, the tuned Lasso struck a balance between bias and variance, effectively reducing overfitting and improving generalization. This demonstrates the value of hyperparameter tuning, as it allowed the model to leverage the benefits of regularization while retaining predictive accuracy. Overall, the tuned Lasso model is the most robust and accurate of the three.

```python
#1
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.feature_selection import SelectFromModel

# Train the Decision Tree with reduced features
clf = DecisionTreeClassifier(max_depth = 3,
                             min_samples_split = 5,
                             random_state = 42,
                             min_samples_leaf = 4
                             )
clf.fit(feat_train, tgt_train)


# Visualize the Decision Tree
plt.figure(figsize=(12, 8))
plot_tree(clf,
          feature_names = features_processed,
          class_names=['0', '1'],
          filled=True,
          fontsize = 10
)
plt.title("Decision Tree for Predicting default")
plt.show()
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_val_score

# Define possible values for hyperparameters
max_depth_values = [3, 5, 10, None]
min_samples_leaf_values = [1, 5, 10]

results = []

# Looping through hyperparameters
for md in max_depth_values:
    for msl in min_samples_leaf_values:

        mod = DecisionTreeClassifier(
            max_depth = md,
            min_samples_split=5,
            random_state=42,
            min_samples_leaf= msl
        )


        scores = cross_val_score(mod, feat_train, tgt_train, cv=5)

        results.append({
            "max_depth": md,
            "min_samples_leaf": msl,
            "mean_accuracy": scores.mean()
        })


results_df = pd.DataFrame(results)

best_result = results_df.loc[results_df["mean_accuracy"].idxmax()]

# Extract the best hyperparameters
best_max_depth = best_result["max_depth"]
best_min_samples_leaf = best_result["min_samples_leaf"]

print(f"Best Hyperparameters: max_depth={best_max_depth}, min_samples_leaf={best_min_samples_leaf}")
print(f"Best Mean Accuracy: {best_result['mean_accuracy']:.4f}")
```

It appears to predict only the majority class, no default, as indicated by the uniform output of 0 and a single color in the visualization. This behavior suggests that the tree is biased toward the majority class because it aims to minimize the overall error, which is dominated by the no default class given the highly imbalanced distribution.
